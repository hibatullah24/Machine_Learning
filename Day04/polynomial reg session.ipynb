# Part1

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(0)
X = np.sort(2 * np.random.rand(100, 1) - 1, axis=0)  
y = 2 * X**2 + X + np.random.randn(100, 1) * 0.1  # Quadratic + noise
```

```python
plt.scatter(X, y, color='blue', label='Data')
plt.title('Synthetic Dataset')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
linear_model = LinearRegression()
linear_model.fit(X, y)

y_pred_linear = linear_model.predict(X)

plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred_linear, color='red', label='Linear Regression')
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
degrees = [1, 2, 5]
plt.scatter(X, y, color='blue', label='Data')

for degree in degrees:
    poly_features = PolynomialFeatures(degree=degree, include_bias=False)
    X_poly = poly_features.fit_transform(X)
    
    poly_model = LinearRegression()
    poly_model.fit(X_poly, y)
    
    y_pred_poly = poly_model.predict(X_poly)
    plt.plot(X, y_pred_poly, label=f'Degree {degree}')

plt.title('Polynomial Regression with Different Degrees')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import SGDRegressor
degrees = [1, 2, 5]
plt.scatter(X, y, color='blue', label='Data')

for degree in degrees:
    poly_features = PolynomialFeatures(degree=degree, include_bias=False)
    X_poly = poly_features.fit_transform(X)
    
    poly_model = SGDRegressor()
    poly_model.fit(X_poly, y)
    
    y_pred_poly = poly_model.predict(X_poly)
    plt.plot(X, y_pred_poly, label=f'Degree {degree}')

plt.title('Polynomial Regression with Different Degrees')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

# Part 2

```python
import numpy as np
from matplotlib import pyplot as plt

from sklearn.linear_model import LinearRegression

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
```

```python
X = 6* np.random.rand(100, 1) - 6
y = 0.3 * X**2 + X + 2 + np.random.randn(100, 1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X, y)
```

```python
model = LinearRegression()
model.fit(X, y)
y_hat = model.predict(X)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X, y)
ax.plot(X, y_hat)
```

```python
X_poly = PolynomialFeatures(3).fit_transform(X)
```

```python
X_poly.shape
```

```python
model2 = Ridge(alpha=0.003)
model2.fit(X_poly, y)
y_hat2 = model2.predict(X_poly)
X_b = np.linspace(X.min(), X.max(), 10).reshape(-1, 1)
X_b_poly = PolynomialFeatures(3).fit_transform(X_b)
y_b_hat = model2.predict(X_b_poly)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X, y)
ax.plot(X_b, y_b_hat, color="red")
ax.set_ylabel("y")
ax.set_xlabel("X")
```

```python
from sklearn.linear_model import SGDRegressor

model =  SGDRegressor()
model.fit(X, y)
y_hat = model.predict(X)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X, y)
ax.plot(X, y_hat)
```
