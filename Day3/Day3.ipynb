```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

y = [85, 45, 75]
y_h = [64, 12, 95]

mse = mean_squared_error(y, y_h)
mae = mean_absolute_error(y, y_h)
print('MSE:', mse)
print('MAE:', mae)
```

```python
y = [85, 45, 75, 738]
y_h = [64, 12, 95, 65]

mse1 = mean_squared_error(y, y_h)
mae1 = mean_absolute_error(y, y_h)

print('MSE:', mse1)
print('MAE:', mae1)
```
