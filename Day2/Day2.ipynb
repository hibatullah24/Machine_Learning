```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
c = np.random.randint(0, 100, size=100)
f = (9/5) * c + 32
```

```python
c.shape, f.shape
```

```python
data = pd.DataFrame({
    "c" : c,
    "f" : f

})
```

```python
data
```

```python
model = LinearRegression()
```

```python
model.fit(c.reshape(-1,1), f.reshape(-1,1))
```

```python
model.coef_, model.intercept_
# it not change even you increase number of columns  
```

```python
#i will add noise to the model is the coef and intercept will change?
for m in [50, 100, 1000, 10000]:
    c = np.random.randint(0, 100, size=m)
    f = (9/5) * c + 32
    f = f + np.random.normal()

    x = c.reshape(-1,1)
    y = f.reshape(-1,1)

    model = LinearRegression()

    model.fit(x,y)
    print(f"m: {model.coef_}, b: {model.intercept_}")
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
X = np.random.randint(0, 100, size=10)
y = np.random.normal() * X + 5 + np.random.randint(0,10) * X
y[6] = y[6] + 10
y[3] = y[3] + -10
y[7] = y[7] + -30
```

```python
plt.scatter(X, y, marker="+")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X.reshape(-1, 1), y.reshape(-1, 1))
```
