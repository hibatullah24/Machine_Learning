# Gradient Descent basic 1

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
import plotly.graph_objects as go
import pandas as pd
from sklearn.metrics import mean_squared_error

import plotly.io as pio
pio.renderers.default = 'iframe'
```

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Visualize the data
plt.scatter(X, y, color='blue', label='Data Points')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Generated Data')
plt.legend()
plt.show()
```

```python
# Standardize the Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Implement Gradient Descent
sgd_reg = SGDRegressor(max_iter=1000, eta0=0.1)
sgd_reg.fit(X_scaled, y.ravel())

# Print the coefficients
print(f"Intercept: {sgd_reg.intercept_}")
print(f"Coefficient: {sgd_reg.coef_}")
```

```python
# Visualize the Regression Line
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, sgd_reg.predict(X_scaled), color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Gradient Descent Fit')
plt.legend()
plt.show()
```

# # gradienr_descent_basics


```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
from matplotlib import pyplot as plt
```

```python
X = 2 * np.random.rand(1000, 1)
y = 4 + 3 * X + np.random.randn(1000, 1) # to add some noise to the data
y_clean = 4 + 3 * X
```

```python
GD_model = SGDRegressor(max_iter=1000, eta0=0.001)
GD_model.fit(X, y.squeeze())
```

```python
y_hat = GD_model.predict(X)
```

```python
GD_model.intercept_, GD_model.coef_
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y, y_hat)
```

```python
GD_model.n_iter_no_change
```

```python
X_b = np.linspace(X.min(), X.max(), 10).reshape(-1, 1)
```

```python
def model_ex(max_iter=0, lr=0, line_style=None):
    _model = SGDRegressor(max_iter=max_iter, eta0=lr)
    _model_fit = _model.fit(X, y)
    _y_hat = _model_fit.predict(X)
    mse = mean_squared_error(y_clean, _y_hat)
    print(f"MSE: {mse}")
    _y_hat_b = _model_fit.predict(X_b)
    line_style["label"] += f" - MSE={mse:0.4f}"
    plt.plot(X_b, _y_hat_b, **line_style)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X, y)
plt.ylabel("y")
plt.xlabel("X")
# # plt.plot(X_b, y_hat_b)
model_ex(5, 0.001, {"color":"red", "label":"lr=0.001 - itter=5"})
model_ex(10, 0.001, {"color":"green", "label":"lr=0.001 - itter=10"})
model_ex(30, 0.001, {"color":"yellow", "label":"lr=0.001 - itter=30"})
model_ex(3000, 0.001, {"color":"black", "label":"lr=0.001 - itter=3000"})
model_ex(4000, 0.001, {"color":"blue", "label":"lr=0.001 - itter=4000"})
plt.legend()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X, y)
plt.ylabel("y")
plt.xlabel("X")
# # plt.plot(X_b, y_hat_b)
model_ex(10, 0.001, {"color":"red", "label":"lr=0.001 - itter=5"})
model_ex(10, 0.01, {"color":"green", "label":"lr=0.001 - itter=10"})
model_ex(10, 0.02, {"color":"yellow", "label":"lr=0.001 - itter=30"})
model_ex(10, 0.05, {"color":"black", "label":"lr=0.001 - itter=3000"})
model_ex(10, 0.5, {"color":"blue", "label":"lr=0.001 - itter=4000"})
plt.legend()
```

# Gradient Descent basic 2


```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error
import pandas as pd

import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
import plotly.io as pio
pio.renderers.default = 'iframe'
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
X = 2 * np.random.rand(1000, 1)
y = 4 + 3 * X 
```

```python
GD_model = SGDRegressor(max_iter=1, eta0=0.0001, warm_start=True)
```

```python
out = []

for i in range(100):
    GD_model.fit(X, y) # Singel itteration of GD
    out.append([
                i, GD_model.coef_[0], # Model Parameter (1)
                GD_model.intercept_[0], # Model Parameter (2)
                mean_squared_error(y, GD_model.predict(X))
                ])
    
    df = pd.DataFrame(out, columns=["itteration", "m", "b", "MSE"])
    _out = out[-1]
    print(f"Itteration Number: {_out[0]}, Param (1): {_out[1]}, Param (2): {_out[2]}, MSE: {_out[3]}")
    
    # input()
    # break
```
