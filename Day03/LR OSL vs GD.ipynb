```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
ols_model = LinearRegression()
```

```python
ols_model.fit(X, y)
```

```python
ols_model.fit(X, y)
```

```python
y_pred_ols = ols_model.predict(X)
print("OLS Intercept:", ols_model.intercept_)
print("OLS Coefficient:", ols_model.coef_)
print("OLS MSE:", mean_squared_error(y, y_pred_ols))
```

```python
sgd_model = SGDRegressor(max_iter=1000, eta0=0.01, learning_rate='constant')
```

```python
sgd_model.fit(X, y.ravel())
```

```python
y_pred_sgd = sgd_model.predict(X)
print("SGD Intercept:", sgd_model.intercept_)
print("SGD Coefficient:", sgd_model.coef_)
print("SGD MSE:", mean_squared_error(y, y_pred_sgd))
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, color='gray', label='Data')
plt.plot(X, y_pred_ols, color='blue', label='OLS Prediction')
plt.plot(X, y_pred_sgd, color='red', linestyle='--', label='SGD Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('OLS vs Gradient Descent')
plt.show()
```
