```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
#change tempreture from c to f
for m in [50, 100, 1000, 10000]:
    c = np.random.randint(0, 100, size=m)
    f = (9/5) * c + 32
    f = f+ np.random.normal()
    
    x = c.reshape(-1,1)
    y = f.reshape(-1,1)

    model = LinearRegression()
    model.fit(x, y)
    print(f"m: {model.coef_}, b: {model.intercept_}")
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
X = np.random.randint(0, 100, size=10)
y = np.random.normal() * X + 5 + np.random.randint(0, 10) * X
y[6] = y[6] + 10
y[3] = y[3] + -10
y[7] = y[3] + -30
```

```python
plt.scatter(X, y, marker="+")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X.reshape(-1, 1), y.reshape(-1, 1))
```

```python
new_X = np.linspace(X.min(), X.max(), 100)
y_hat = model.predict(new_X.reshape(-1, 1))
```

```python
plt.scatter(X, y, color="red",marker="+")
plt.plot(new_X, y_hat)
```
